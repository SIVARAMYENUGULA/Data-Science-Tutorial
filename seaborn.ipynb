```python
print("---------------------------pair plots-----------------------")
import seaborn
import matplotlib.pyplot as plt
df = seaborn.load_dataset("tips")
print(df.to_string())
seaborn.pairplot(df)
plt.show()
```

```python
print("---------------------------plotting selected variables-----------------------")
import seaborn
import matplotlib.pyplot as plt
df = seaborn.load_dataset("tips")
selected_variables = df[["total_bill", "tip"]]
seaborn.pairplot(selected_variables)
plt.show()
```

```python
print("---------------------------adding a hue color-----------------------")
seaborn.pairplot(df, hue="size")
plt.show()
```

```python
print("---------------------------changing the color palette-----------------------")
seaborn.pairplot(df, hue="size",palette="husl")
plt.show()
```

```python
print("---------------------------changing the diagonal plot-----------------------")
seaborn.pairplot(df, diag_kind="kde")
plt.show()
```

```python
print("---------------------------adjusting the kind of plot-----------------------")
seaborn.pairplot(df, kind="reg")
plt.show()
```

```python
print("---------------------------controlling the markers-----------------------")
seaborn.pairplot(df, hue="sex",markers=["o","s"])
plt.show()
```

```python
print("---------------------------advanced customization-----------------------")
g=seaborn.pairplot(df,hue="day")
g.fig.suptitle("Pairplot of tips dataset",y=1.02)
g.set(xticks=[10],yticks=[20])
plt.show()
```
